Make a new dataframe called X that subsets the Log GDP per capita through and including Confidence in national government.

```python
X= df[['Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices',
       'Generosity','Perceptions of corruption','Positive affect','Negative affect','Confidence in national government']]
X
```

Subset df['LifeLadder] from df and assign to a variable called y.

```python
y = df[['Life Ladder']]
y
```

Convert X and y to numpy arrays. Split your data in X_train, X_test, y_train, y_test using 7 digit as the random seed and an 80/20 split.

```python
print(X.shape, y.shape)
```

```python
#importing sklearn functions
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
#Spliting the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=3003795) #my student id is 3003795
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
X_train
```

```python
#Converting the X & y variables to numpy arrays
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

Fitting a linear regression model. Calculating mean absolute error for each partition. Making a scatterplot of actual vs. predicted values for each partition with nice axis labels, titles and markers.

```python
#Fitting a regression model
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
tmp = pd.DataFrame(X_train)  # converting to pd to run the statistics for X_train
tmp.describe() #max values are 1, min values are 0
```

```python
tmp = pd.DataFrame(X_test)  #converting to pd to run the statistics for X_test
tmp.describe() #max values are 1, min values are 0
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression() #defining the variable to store the linear regression model
LR = LR.fit(X_train, y_train) 
```

```python
#Storing the predictions into train_pred and test_pred variables

train_pred = LR.predict(X_train)
test_pred = LR.predict(X_test) 
```

```python
from sklearn.metrics import r2_score

# Calculating R2 value for actual and predicted values

print("The train R2 value is:", r2_score(y_train, train_pred))
print("The test R2 value is:", r2_score(y_test, test_pred))

```

```python
from sklearn.metrics import mean_absolute_error

# Calculating mean absolute error for each partition

trainMAE = mean_absolute_error(y_train, train_pred)
print("The train MAE value is:", trainMAE)
testMAE = mean_absolute_error(y_test, test_pred)
print("The test MAE value is:", testMAE) 
```

```python
#Plotting a scatter plot for train results

plt.figure(figsize=(8, 6))
plt.scatter(x=y_train, y=train_pred,color='r')
plt.plot([0, 10], [0, 10], '--k') #45 degree line
plt.axis('tight')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.suptitle('Train Results for Actual Vs Predicted values')
plt.show()

```

```python
#Plotting a scatter plot for test results

plt.figure(figsize=(8, 6))
plt.scatter(x=y_test, y=test_pred,color= 'blue')
plt.plot([0, 10], [0, 10], '--k') # 45 degree line
plt.axis('tight')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.suptitle('Train Results for Actual Vs Predicted values')
plt.show()
```

Writing ten-fold cross-validation on X_train with a for loop. Apply the model each of the ten times to the X_test partition (once for each fold). Store results efficiently in the for loop! Calculating mean absolute error for each fold and application of the model to X_test. 
Plotting a scatterplot of actual vs. predicted values for each fold and for each application of the model to X_test with nice axis labels, titles and markers. 

```python
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
y_train = pd.DataFrame(y_train)
y_train = pd.DataFrame(y_train)
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import RepeatedKFold
from sklearn.utils import shuffle


foldRange = np.arange(10)
repeatRange = np.arange(10)
resultDF = []

for a in repeatRange: 
  X_train, y_train = shuffle(X_train, y_train) # shuffles them togethe
  for i in foldRange:
    model_X_test = X_train[round(i*0.1*X_train.shape[0]) : round((i+1)*0.1*X_train.shape[0])]
    model_y_test = y_train[round(i*0.1*y_train.shape[0]) : round((i+1)*0.1*y_train.shape[0])]
    model_X_train = X_train[~X_train.isin(model_X_test)].dropna()
    model_y_train = y_train[~y_train.isin(model_y_test)].dropna()
    LR = LR.fit(model_X_train, model_y_train)                                 
    test_preds = LR.predict(model_X_test)
    test_MAE = mean_absolute_error(model_y_test, test_preds) # to calculate MSE
    resultDF.append([i,test_MAE])#append it  

# we need three columns not all in one list, and we need to select 
# rows in order after shuffling
```

```python
resultDF = pd.DataFrame(resultDF)
resultDF
```
